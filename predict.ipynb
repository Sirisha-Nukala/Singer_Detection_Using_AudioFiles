```python
import librosa
import numpy as np

def extract_features(file_path):
    # Load the audio file
    y, sr = librosa.load(file_path, duration=30) # load up to 30 seconds
    
    # 1. Extract 13 MFCCs
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    mfccs_mean = np.mean(mfccs.T, axis=0) # Get the mean of each of the 13 coefficients
    
    # 2. Extract Spectral Centroid
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)
    centroid_mean = np.mean(spectral_centroid)
    
    # 3. Extract Zero Crossing Rate (ZCR)
    zcr = librosa.feature.zero_crossing_rate(y)
    zcr_mean = np.mean(zcr)
    
    # 4. Extract Chroma Mean
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    chroma_mean = np.mean(chroma)
    
    # Combine everything into a single list of 16 features
    features = list(mfccs_mean) + [centroid_mean, zcr_mean, chroma_mean]
    
    # Reshape for the model (1 sample, 16 features)
    return np.array(features).reshape(1, -1)

import joblib

def predict_singer(file_path):
    # Load the saved model and encoder
    model = joblib.load('svc_singer_model.pkl')
    encoder = joblib.load('label_encoder.pkl')
    
    # Extract features from the uploaded audio file
    features = extract_features(file_path)
    
    # Make a prediction (returns a number)
    prediction_num = model.predict(features)
    
    # Convert the number back to the singer's string name
    singer_name = encoder.inverse_transform(prediction_num)
    
    return singer_name[0]

# Test it locally:
print("Analysed Singer:")
print(predict_singer("VocalSet11/FULL/female2/excerpts/straight/f2_dona_straight.wav"))
```
